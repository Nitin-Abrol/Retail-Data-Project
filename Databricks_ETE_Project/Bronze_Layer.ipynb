### **Data Reading**

```python
df = spark.readStream.format("cloudFiles")\
                  .option("cloudFiles.format","parquet")\
                  .option("cloudFiles.schemaLocation",f"abfss://bronze@storageaccnitin.dfs.core.windows.net/checkpoint_{file_name}")\
                  .load(f"abfss://source@storageaccnitin.dfs.core.windows.net/{file_name}")
          
```

```python
df.writeStream.format("parquet")\
    .outputMode("append")\
    .option("checkpointLocation",f"abfss://bronze@storageaccnitin.dfs.core.windows.net/checkpoint_{file_name}")\
    .option("path",f"abfss://bronze@storageaccnitin.dfs.core.windows.net/{file_name}")\
    .trigger(once=True)\
    .start()
```
